```python
def func(x):
    return x**4 - 2*x + 1

def Simpson(a,b,N):
    I = func(a) + func(b)
    h = (b-a)/ N
    for k in range(1,N,2):
        I += 4*func(a + k*h)
    for k in range(2,N,2):
        I += 2*func(a + k*h)
    I = I/(3.0)
    I = I*h
    return I
print(Simpson(0,2,10))
print(Simpson(0,2,100))
print(Simpson(0,2,1000))
```

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def CV(T):
    V = 1000
    kb = 1.38e-23
    rho = 6.022e28
    Debye = 428
    N = 1000
    
    def f(x):
        return (x**4 * np.exp(x))/(np.exp(x)-1)**2
    
    c = V*rho*kb*(T/Debye)**3
    b = Debye/T
    h = (b-0)/N
    cv = .5*0 + .5*f(b)
    
    for k in (1,N,1):
        cv += f(0 + k*h)
    return cv*h*9*c

Temperature = []
cv = []

for v in range(1, 500):
    Temperature.append(v)
    cv.append(CV(v))

    
fig = plt.figure(figsize=(20,15))
ax =  fig.add_subplot(1,1,1)
image = ax.scatter(Temperature,cv,s = 100,edgecolor='k',alpha=0.5)

ax.set_xlabel("$Temperature$ $(K)$",size=15)
ax.set_ylabel("$Heat Capacitance$",size=15)

```
