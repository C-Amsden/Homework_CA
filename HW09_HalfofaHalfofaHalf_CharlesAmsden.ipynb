```python
import random as ran
from numpy import arange
from matplotlib import pyplot as plt


#Starting Conditions
Bi_213 = 10000
Tl_209 = 0
Pb_209 = 0
Bi_209 = 0

#constants
tau_Bi = 46.0*60 #to make it seconds
tau_Tl = 2.2*60
tau_Pb = 3.3*60
dt = 1.0 #second step size
p_Bi = 1 - 2**(-dt/tau_Bi) #Probablity of decay in that second
p_Tl = 1 - 2**(-dt/tau_Tl)
p_Pb = 1 - 2**(-dt/tau_Pb)

time = arange(0.0,20000,dt)
Bi_213count = []
Tl_209count = []
Pb_209count = []
Bi_209count = []

for t in time:
    Bi_213count.append(Bi_213)
    Tl_209count.append(Tl_209)
    Pb_209count.append(Pb_209)
    Bi_209count.append(Bi_209)
    
    li_1 = ['pb','tl']
    li_2 = [.9791, .0209]
    
    random_dist = ran.choices(li_1,li_2)
    if random_dist is 'pb':
        decay = 0
        for x in range(Bi_213):
            if ran.random() < p_Bi:
                decay += 1
        Bi_213 -= decay
        Pb_209 += decay
    else:
        decay = 0
        for x in range(Bi_213):
            if ran.random() < p_Bi:
                decay += 1
        Bi_213 -= decay
        Tl_209 += decay
     
    decay = 0
    for x in range(Tl_209):
        if ran.random() < p_Tl:
            decay += 1
    Tl_209 -= decay
    Pb_209 += decay
    
    decay = 0
    for x in range(Pb_209):
        if ran.random() < p_Pb:
            decay += 1
    Pb_209 -= decay
    Bi_209 += decay


plt.plot(time,Bi_213count, c='y')
plt.plot(time,Tl_209count, c='g')
plt.plot(time,Pb_209count, c='b')
plt.plot(time,Bi_209count, c='r')
plt.xlabel('Time in Seconds')
plt.ylabel('Counts')
plt.show()
#plt.savefig('Half_lifeHW09.png')
```
