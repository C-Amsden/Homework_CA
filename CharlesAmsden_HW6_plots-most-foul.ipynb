```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
%matplotlib inline

data = np.genfromtxt("CG20.dat", delimiter=",", names = True)


fig = plt.figure(figsize=(20,15))

ax1 =  fig.add_subplot(2,2,3)
image = ax1.scatter(data['Rgc'],data['Z'],s = 200*data['r50'], c = data['AgeNN'],edgecolor='k',cmap="plasma",alpha=0.5)
bar = fig.colorbar(image,orientation="vertical",pad=0.001)
bar.set_label('$log_{10} (Age)$',size=15)

ax1.set_xlabel("$R_{GC}$ $(pc)$",size=15)
ax1.set_ylabel("$Z_{GC}$ $(pc)$",size=15)


ax2 =  fig.add_subplot(2,2,4)
image = ax2.scatter(data['AgeNN'],data['DistPc'],s = 200*data['r50'], c = data['AgeNN'],edgecolor='k',cmap="tab10",alpha=0.5)
bar = fig.colorbar(image,orientation="vertical",pad=0.001)
bar.set_label('$log_{10} (Age)$',size=15)

ax2.set_xlabel("$AgeNN$",size=15)
ax2.set_ylabel("$Dist$ $(pc)$",size=15)

ax3 =  fig.add_subplot(2,2,1)
image = ax3.scatter(data['GLON'],data['GLAT'],s = 200*data['r50'], c = data['AgeNN'],edgecolor='k',cmap="YlGn",alpha=0.5)
bar = fig.colorbar(image,orientation="vertical",pad=0.001)
bar.set_label('$log_{10} (Age)$',size=15)

ax3.set_xlabel("$GLON$",size=15)
ax3.set_ylabel("$GLAT$",size=15)

ax4 =  fig.add_subplot(2,2,2)
image = ax4.scatter(data['pmRA'],data['pmDE'],s = 200*data['r50'], c = data['AgeNN'],edgecolor='k',cmap="binary",alpha=0.5)
bar = fig.colorbar(image,orientation="vertical",pad=0.001)
bar.set_label('$log_{10} (Age)$',size=15)

ax4.set_xlabel("$pmRA*$",size=15)
ax4.set_ylabel("$pmDE$",size=15)
ax4.set_xlim(20,-15)
ax4.set_ylim(20,-15)

#ax_sub = fig.add_axes()
#ax_sub.hist(data['nbstars07'])
```
