```python
import random as ran
from numpy import arange
from pylab import plot,xlabel,ylabel,show

def ThreeDoorsGame(t):
    results_switches = []
    results_stays = []
    doors = []
    y = 0
    for x in range(t):
        doors.append(ran.randint(0,1))
        if sum(doors) == 0:
            doors.append(ran.randint(0,1))
            if sum(doors) == 0:
                doors.append(1)
            else:
                doors.append(0)
        else:
            doors.append(0)
            doors.append(0)
        
        
        
        pick = ran.randrange(0,3,1)
        
    
    
        if pick == 2:
            if doors[0] ==0:
                doors.pop(0)
            else:
                doors.pop(1)
        elif pick == 1:
            if doors[0] == 0:
                doors.pop(0)
            else:
                doors.pop(2)
        else:
            if doors[1] == 0:
                doors.pop(1)
            else:
                doors.pop(2)
        
        if pick == 2:
            pick -= 1
        
        #print(doors)
        #print("pick: ",doors[pick])
        
        if pick == 0:
            switch = doors[1]
        else:
            switch = doors[0]
        
        win = doors.index(1)
        
        if y < t/2:
            pick = doors.index(switch)
            #print(switch)
            #print("switch: ", doors[pick])
            if doors[pick] == doors[win]:
                results_switches.append("win")
                #print('win')
    
            else:
                results_switches.append("lose")
                #print('lose')
    
    
        else:
            pick = pick
            #print("stay: ", doors[pick])
            if doors[pick] == doors[win]:
                results_stays.append("win")
    
            else:
                results_stays.append("lose")
    
        
    
            
        #if doors[pick] == doors[win]:
            #results_switches.append("win")
        #else:
            #results_switches.append("lose")
        
        y += 1
    
        doors = []
    #print(results_stays.count('win'),len(results_stays))
    #print(results_switches.count('win'),len(results_switches))
    i,tot_st = results_stays.count('win'),len(results_stays)
    r, tot_sw = results_switches.count('win'), len(results_switches)
    print('probablity of winning if you switch: ', (r/tot_sw)*100, '%')
    print('probablity of winning if you stay: ', (i/tot_st)*100,'%')
    print('the probablity of improving by switching(about): ', (r-i)/(t/2)*100, '%')
```
